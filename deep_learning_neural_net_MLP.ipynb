**Neural Networks & Multi-Layer Perceptron**

## Neural Networks
Neural networks: 애매하게 뇌를 모방한 기술. (뉴런의 신경전달을 약간 모방)<br>
예를 들어, 하늘을 날 때, 처음엔 새를 모방했다가, 라이트형제가 프로펠러가 달린 비행기를 만들고, 요즘은 전투기가 나온다.<br>
전투기는 새와 하나도 닮지 않았다. 이런것처럼 뉴럴 네트워크도 뇌와 같은게 아니고, 신경전달 아이디어만 차용한 수학에 가깝다.<br>

### Linear Neural Networks
가장 간단한 뉴럴 네트워크는 선형 회귀. -> y = wx+ b에서 예측y와 실제y 차이가 가장 작은 w와 b를 찾는 것(경사하강법 이용)<br>
w와 b를 각각 편미분해서 학습율(Stepsize)에 곱해 더하는 것(편미분 결과 음수기 때문에 뺴는것처럼 보임)이 Loss Function을 최소화 하는 방향<br>

### Beyond Linear Neural Networks
Nonlinear transform(비선형 변환)이 필요하다.<br>
이게 활성 함수(Activation fucntion)이다. -> 이게 존재하는 여려층의 인공신경망을 Multi-Layer Perceptron(MLP)라고 한다.

### Loss Function
Loss Function에 따라, 딥러닝의 학습 방향이 많이 달라지게 된다.<br>
예를 들어, 아웃라이어가 큰 데이터셋에 대해 MSE를 하면 아웃라이어 오류 차이가 제곱이 되어, 이를 낮추려다보니 아웃라이어에 오버피팅되느 모델이 이상해질 수 있다.<br>

1. MSE : 2-norm기반 오류의 합.
2. CE(Cross Entrophy): 분류문제 풀 때 좋다.
3. MLE(확률) : 회귀할 때, 사진을 보고 20대일 확률, 30대일 확률 등 전체적인 확률을 보고싶은 경우 사용
