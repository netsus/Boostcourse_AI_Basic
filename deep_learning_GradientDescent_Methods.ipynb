## Gradient Descent의 분류 (3가지)
1. SGD(Stochastic gradient descent): 한 번에 1개씩 GD를 구해서 업데이트하는 방식.<br>
<br>
2. Mini-batch gradient descent: BatchSize만큼에 대해 GD를 구해서 업데이트하고, 다음 Batch에 대해서도 반복.<br>
(전체 데이터를 Batch로 다 돌때까지 반복한게, 한번 GD 업데이트 한 것.)<br>
<br>
3. Batch gradient descent: 한번에 전체 데이터의 모든 GD를 구해서 업데이트.<br>
<br>
-> 대부분의 딥러닝은 Mini-batch를 사용한다.

### Batch-size 문제
배치 사이즈가 작을 수록, Flat Minimum에 도달하고, 배치 사이즈가 클 수록 Sharp Minimum에 도달한다.<br>
Flat Minimum에서는 테스트 데이터에 대한 성능이 약간 떨어지지만,<br>
Sharp Minimum에선 테스트 데이터에 대한 성능이 엄청 안좋아질 수 있다. 그 이유는, 약간만 비껴가도 성능이 급변하기 때문.<br>
즉, Flat Minimum의 Generalization 성능이 더 좋다. -> 배치 사이즈가 작을 수록 Generalization 성능이 좋아진다.<br>

### Momentum
모멘텀은 베타가 모멘텀(관성)을 일으키는 특징을 갖는다.<br>
즉, 한번 흘러간 방향에 대한 가중치를 부여.<br>

### NAG
Gradient를 계산할 때, Lookahead gradient라는 것을 계산한다.<br>
모멘텀은 관성 때문에 로컬 미니엄 주변에서 수렴하지 못하고 왔다갔다 할 가능성이 있다.<br>
그에 반해, NAG는 로컬 미니멈에 빠르게 수렴하는 가중치를 준다.<br>
<br>

### Adagrad
파라미터들이 지금까지 얼마나 변해왔는지, 안변해왔는지를 보고,<br>
많이 변한 파라미터는 덜 변하게 하고, 적게 변한 파라미터는 많이 변하게 한다.<br>
이때, 파라미터가 얼마나 변했는지를 제곱해서 더한게 G이다.<br>
G는 계속 커지기 때문에, 그 역수는 0에 수렴하고, 학습이 멈추는 경향이 생긴다.<br>
<br>

### Adadelta
Adagrad에서 가지고있던, learning rate이 Gt의 역수로 표현됨으로써 생기는 문제(학습이 멈추는 문제)를 막는 방법.<br>
Adadelta는 learning_rate이 없기 때문에 잘 안쓴다.<br>
