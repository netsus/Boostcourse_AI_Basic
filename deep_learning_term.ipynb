### Loss
실제값과 예측값의 차이인 로스를 줄여야 성능이 좋다고 할 수 있지만, 로스 펑션은 이루고자 하는것의 근사치임을 명심해야한다.<br>
여러 종류의 로스펑션이 있고, 문제에 따라 적용이 다 다르게되는 특징을 이해해야 한다.<br>

    예를 들어, 회귀에서 아웃라이어 때문에 MSE가 값이 클 때, 다른 MSLE나 MLE 등을 써서 성능을 올릴 수 있다.<br>
    -> 풀고자 하는 문제와 Loss Function과의 관계를 정확히 아는게 중요하다.<br>
    목적은 학습하지 않은 데이터에서 모델이 잘 동작하는것이다!

### Algorithm
최적화에 사용되는 SGD(Stochastic Gradient Descent) 외에도 다양한 알고리즘이 있다.<br>
이런 최적화 알고리즘 뿐만 아니라 성능을 향상시키기 위해, <br>
Dropout, Early Stopping, k-fold, Baysian Optimization, Weight decay 등등의 알고리즘적 방법을 쓴다.<br>
목표는 학습하지 않은 데이터에서, 최고의 성능을 내기 위한것!
