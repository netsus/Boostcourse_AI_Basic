### Generalization
학습데이터를 학습할 때, iterate가 증가하면 Error가 0에 가깝게 줄어든다.<br>
하지만, 이것이 테스트 데이터에 대한 성능을 보장해주지 않는다. 이 사이 차이가 있는데 이를 Generalization gap이라고 한다.<br>
논문에서 이 모델은 좋은 Generalization 성능을 갖는다 라고하면, 성능이 테스트 데이터에 대해 학습 데이터와 비슷하게 나오는것을 어느정도 보장해준다는 것을 의미한다.

### Underfitting vs Overfitting
오버피팅: 학습데이터에 대해선 다 잘맞추는데, 모델이 뭔가 조잡하다. -> 테스트데이터 성능이 떨어진다.<br>
언더피팅: 모델이 너무 간단해서, 학습데이터도 못맞추고, 테스트 데이터도 못맞춘다.<br>
그 사이 어딘가 스윗-스팟(Balanced)가 존재할거라는 이론적 예상이 있다. <br>

### Cross-validation
K-fold validation이라고 불리기도 한다.<br>
학습데이터에 대해서, 여러 폴드로 나누어 1개씩 돌아가며 Validation data로 쓰는 것.<br>
일반 적으로 Cross-Validation을 써서, 최적 하이퍼파라미터를 찾고, 해당 하이퍼 파라미터로 전체 데이터를 학습에 쓴다.<br>
**이유는 Cross-Validation을 하면, 1개의 Chunk를 계속 Validation하는데 쓰니까 학습 데이터개수가 줄어들기 때문이다.**

### Bias and Variance
Variance: 출력이 얼마나 일관적으로 나오는지를 의미 (분산)<br>
Bias: 예측값들이 타겟값과 얼마나 가까운지를 의미

1. Low Variance: 출력이 얼마나 일관적으로 나오는지(분산이 작은 경우)<br>
-> 간단한 데이터셋에서 많이 보인다.<br>
2. High Variance: 비슷한 입력이 들어와도 출력이 많이 달라진다. 일관적이지 않다. (분산이 큰 경우)<br>
-> 복잡한 데이터셋에 서 많이 보인다.<br>
3. Low Bias: 예측값들이 타겟값과 가까운 경우<br>
4. High Bias: 예측값들이 타겟값과 먼 경우

#### Bias and Variance Tradeoff
내 학습데이터에 노이즈가 껴있다고 가정했을 때, 2Norm 기준 에러를 미니마이즈 하는것은 3가지 파트로 나뉠 수 있다.<br>
즉, cost를 미니마이즈 하는것은 bias, variance, noise를 미니마이즈한다는 것으로 나뉜다.<br>

bias를 낮추면 variance가 커지고, variance를 낮추면 bias가 커진다 -> tradeoff<br>
즉, bias, variance를 동시에 낮추는건 근본적으로 어렵다는걸 보여주는 개념<br>

### Bagging vs Boosting
 1) Bagging: Bootstrapping aggregating의 줄임말. 학습 데이터를 여러 랜덤서브샘플로 나누어 여러 모델들로 학습하고 합치는것을 의미.<br>
여러 모델의 예측 결과를 평균을 내서 최종 예측 결과 생성<br>
 2) Boosting: 여러 모델들을 Sequential하게 합쳐서, 하나의 강한 모델을 만드는 것.
