## RNN(Recurrent Neural Network)
위에서 시간의 흐름에 따라 데이터 입력받는걸 가장 쉽고 잘 구현한게 RNN이다<br>
위의 모델과 차이점은, 각 뉴런별로 자기자신에게 돌아오는 연산이 있다는 점이다.<br>
풀어쓰면 X0 X1 X2 .. .Xt 까지 시간순으로 입력받는 인공신경망이다.<br>

### RNN의 단점
Short-term dependencies는 잡지만, Long-term dependencies는 못잡는 문제<br>
과거에 얻어진 어떤 정보들이 다 취합되서 미래에 그게 고려가 되야하는데, RNN은 fix된 rule로 지속적으로 취합하기 때문에, 먼 과거의 정보가 미래까지 살아남기 힘들다.<br>
즉, 현재로부터 적당한 과거의 데이터는 고려가 잘되는데, 먼 과거의 데이터는 희석되서 없어져버리는 문제<br>
마치 메멘토처럼 짧은것만 기억하는 느낌.<br>
즉, RNN은 Short-term dependencies는 잡지만, Long-term dependencies는 못잡는 문제가 발생한다.<br>

## LSTM
위와 같은 Short-term dependencies는 잡지만, Long-term dependencies는 못잡는 문제를 극복하고자 나온게 LSTM이다.<br>
핵심은 어떻게 Short-term dependency를 해결했는지 이다.<br>
LSTM의 단위를 보면, 3개의 input(Previous cell state, Previouse hidden statr, input)이 들어오고, 실질적으론 1개의 Ouput(hidden state)이 나가고, 2개의 Output(Next cell state, Next hidden state)는 다음 LSTM 단위에 전달된다.<br>

### LSTM의 코어 아이디어
중간에 흘러가는 Cell state선이 핵심 아이디어이다. 이 선으로 인해 과거의 데이터들이 잘 요약되어 지속적으로 전달된다.<br>
<br>

### LSTML의 게이트1
1. Forget Gate: 어떤 정보를 버릴지 결정. 현재 입력과 과거의 hidden state를 입력받아서 가중치 곱하고, 활섬하수 통과해서 웨이트 조절하여 버릴애들을 결정.
2. Input Gate: 현재 입력으로 받은 정보중에 어떤 정보를 저장할지 결정. 

<br>
<div>
    <img src=attachment:image.png width=800>
</div>

### LSTM의 게이트2
3. Update cell: Forget Gate와 Input Gate를 통과한 내용을 바탕으로 cell state를 업데이트하여 다음 LSTM 단위에 전달.
4. Output Gate: 업데이트된 cell state를 한번더 조작해서, 출력할 값을 Output(hidden state)로 내보내고, 다음 셀의 히든 스테이트로도 들어간다.

## GRU(Gated Recurrent Unit)
게이트가 2개밖에 없다. reset gate, update gate<br>
reset gate: forget gate와 비슷한 역할. update gate역할이 합쳐짐.<br>
똑같은 모델에 대해서, LSTM보다 GRU를 활용할 때 성능이 올라간다. 이유는 같은 기능을 수행하는데 파라미터 개수가 더 작으니까 Generalization 성능이 더 올라가기 때문이다.<br>
