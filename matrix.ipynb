## 행렬은 뭔가요?
행렬(matrix)은 벡터를 원소로 가지는 2차원 배열입니다.<br>
-> **행벡터**를 원소로 갖는 2차원 배열이다
<br>
<div>
    <img src=attachment:image.png width=800>
    * 행렬의 연산: 전치행렬 -> 행과 열이 바뀐다.<br>
    * 벡터에도 전치연산이 적용된다. 행벡터를 전치하면 열벡터가되고, 열벡터를 전치하면 행벡터가된다.
</div>

### 행렬을 이해하는 방법 1
- 벡터가 공간에서 한 점을 의미한다면, 행렬은 행벡터를 한점으로 하는 여러 점들의 배열이다.
<br>
<div>
    <img src=attachment:image.png width=800>
</div>

### 행렬의 덧셈, 뺄셈, 성분곱, 스칼라곱
1. 행렬은 같은 모양을 가지면 덧셈, 뺄셈이 가능하다.<br>
-> 각 성분간 덧셈과 뺄셈(벡터와 동일)

2. 성분곱 역시 같은 모양끼리 가능하다<br>
-> 각 성분간 곱셈 (벡터와 동일)

3. 스칼라곱은 각 성분에 스칼라값을 곱해주는 것이다<br>
-> 이 역시 벡터와 동일

### 행렬 곱셈
행렬 곱셈은 i번쨰 행벡터와 j번째 열벡터 사이의 내적을 성분으로 가지는 행렬을 계산.<br>
행렬곱은 X의 열의 개수와 Y의 행의 개수가 같아야 한다.
<br>
<div>
    <img src=attachment:image.png width=800>
</div>

```python
import numpy as np
# 일단 seed 값을 지정하고, 
np.random.seed(seed=12) 
X = np.random.randint(1,4,size=9).reshape(3,3)
Y = np.random.randint(2,5,size=9).reshape(3,3)
print(f"{X=}\n{Y=}\n")

## 행렬 곱셈
print(f"{X@Y=}")
```

```python
## 행벡터 끼리의 행렬곱
## -> 두 행렬사이의 열개수가 같아야 계산이 가능하다.
print(f"X와 Y의 np.inner\n{np.inner(X,Y)}\n")
print(f"X와 Y의 행벡터 끼리의 행렬 곲은 X @ Y.T와 동일하다\n{X @ Y.T}")
```
