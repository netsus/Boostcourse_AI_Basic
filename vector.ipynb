## 벡터가 뭔가요?
벡터는 숫자를 원소로 가지는 리스트 또는 배열입니다.<br>
ex) x = [1, 7, 2]
<br>
- 벡터의 종류: 열벡터, 행벡터
<div>

</div>
- 벡터는 공간에서 한 점을 나타낸다. (1차원 벡터는 수직선에서 한 점, 2차원 벡터는 좌표평면에서 한 점)<br>
- 벡터는 원점으로부터 상대적 위치를 표현 (원점 - 공간의 한점)
- 벡터에 숫자(스칼라값)을 곱해주면 길이만 변한다.
- 벡터끼리 곱셈 -> 성분 곱

### 벡터의 norm
1. 정의: 원점에서부터의 거리
2. N차원 벡터에 대해 모두 성립한다. ||x||1과 ||x||2 처럼 다양한 norm이 존재
- L1 norm (||x||1): 각 성분의 절대값을 모두 더해준 것. 좌표평면에서 각 성분의 길이를 모두 더해준것.

- L2 norm (||x||2): 유클리드 거리를 의미

```python
# 벡터의 norm
def l1_norm(x):
    x_norm = np.abs(x)
    x_norm = np.sum(x_norm)
    return x_norm

def l2_norm(x):
    x_norm = x**2
    x_norm = np.sum(x_norm)
    x_norm = np.sqrt(x_norm)
    return x_norm
```

### 두 벡터 사이의 각도 구하기
: cos 2법칙 -> 내적 (꼭 2,3차원 아니고 n차원이더라도 각도 구할 수 있다.)<br>
단, 각도는 항상 l2_norm으로만 구할 수 있다!<br>
<div>

</div>

```python
def angle(x,y):
    v = np.inner(x,y) / (l2_norm(x)*l2_norm(y)) # np.inner는 내적
    theta = np.arccos(v) # v는 코사인값 -> arccos로 각도구할 수 있다.
    return theta
```

### 벡터의 내적
: 내적은 정사영(orthogonal projection)된 벡터의 길이와 관련 있다.<br>
내적은 x를 y에 정사영한 뒤, 길이를 y만큼 곱해준 것이라고 보면된다.<br>
-> 두 벡터간 유사도를 파악할 때 내적값을 쓴다.
